```python
encode_AgeCategory = {'55-59':57, '80 or older':80, '65-69':67,
                      '75-79':77,'40-44':42,'70-74':72,'60-64':62,
                      '50-54':52,'45-49':47,'18-24':21,'35-39':37,
                      '30-34':32,'25-29':27}
df['AgeCategory'] = df['AgeCategory'].apply(lambda x: encode_AgeCategory[x])
df['AgeCategory'] = df['AgeCategory'].astype('int')

#Mudança da variável idade de uma variável categorica para uma variável inteira, que reflete a média das categorias.
```

```python
fig, x = plt.subplots(figsize = (14,6))
sns.kdeplot(df[df["HeartDisease"]=='Yes']["AgeCategory"], alpha=1,shade = False, color="blue", label="HeartDisease", x = x)
x.set_xlabel("Idade")
x.set_ylabel("Frequencia")
plt.show()



#Quanto maior a idade, maior a chance de se ter doenças cardiovasculares.
```

```python
fig, x = plt.subplots(figsize = (8,5))

x.hist(df[df["HeartDisease"]=='No']["Sex"], color="Green")
x.hist(df[df["HeartDisease"]=='Yes']["Sex"], color="Red")

x.set_xlabel("Sexo")
x.set_ylabel("Frequência")


#Homens são mais propenso a ter doenças cardiovasculares.
```
